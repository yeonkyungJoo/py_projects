```python
# 연속된 날짜 리스트 획득
import pandas as pd
```

```python
# (날짜 계산기로 오늘 날짜까지 period 값 계산)
dates = pd.date_range('2005.02.07', periods=5297, freq='D')
# dates
dates[-2 :]   # 슬라이싱
```

```python
# 변화하는 값이 맨 뒤에 있으므로 문자열 '이어붙이기'
target_site = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date='
for date in dates[-2 :] :
    # 필요한 형태는 YYYYmmdd
    # print(date, type(date), date.strftime('%Y%m%d') )
    print(target_site + date.strftime('%Y%m%d'))
```

```python
for date in dates :
    print(target_site + date.strftime('%Y%m%d'))
```

```python
# 진행율 표시 : tqdm
# > conda install tqdm
```

```python
from tqdm import tqdm
```

```python
import time
tmp = []
for date in tqdm(dates) :
    tmp.append(target_site + date.strftime('%Y%m%d'))
    time.sleep(1*0.1) # 0.1초
```

```python
for date in tqdm(dates[-2: ]) :   # (데이터가 너무 많아서 슬라이싱)
    page = urlopen(target_site)
    soup = BeautifulSoup( page, 'html.parser' )
    # 영화명 (tit5 클래스)
    titles = soup.select('.tit5') 
    # 평점 (point 클래스)
    points = soup.select('.point')
    # 날짜
    print(titles, points, date)
```

```python
for date in tqdm(dates[-2: ]) :   # (데이터가 너무 많아서 슬라이싱)
    page = urlopen(target_site)
    soup = BeautifulSoup( page, 'html.parser' )
    # 영화명 
    for title in soup.select('.tit5') :
        print(title.a.string.strip())
    # 평점 
    for point in soup.select('.point') :
        print(point.string.strip())
```

```python
# 영화명, 평점 같이 출력
for i in range( len(soup.select('.tit5'))) :
    print(titles[i].a.string.strip())
    print(points[i].string.strip())
```

```python
# [ { 'time' : xx, 'movie' : xx, 'point' : 3.89}, {}, {} ]  형태로 출력
results = [
    {
        'time' : date,
        'movie' : titles[i].a.string.strip(),
        'point' : points[i].string.strip()
    }
    for i in range( len(soup.select('.tit5')))
]
```
