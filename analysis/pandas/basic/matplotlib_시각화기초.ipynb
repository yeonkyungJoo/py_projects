```python
import numpy as np
```

```python
# 0 ~ 12 사이를 0.01 간격으로 쪼개라
x = np.arange(0, 12, 0.01)
x
```

```python
y = np.sin(x)
plt.figure()
plt.plot(x, y)
plt.show()
```

```python
plt.figure( figsize = (10, 6) )
plt.plot(x, y)
# 격차 배치
plt.grid()
plt.show()
```

```python
plt.figure( figsize = (10, 6) )
plt.plot(x, y)
plt.grid()
# 라벨
plt.xlabel('time')
plt.ylabel('값')
plt.show()
```

```python
plt.figure( figsize = (10, 6) )
plt.plot(x, y)
plt.grid()
plt.xlabel('time')
plt.ylabel('값')
# 제목
plt.title('sin 함수 시각화')
plt.show()
```

```python
plt.figure( figsize = (10, 6) )
plt.plot(x, y)
# 데이터 추가
plt.plot(x, np.cos(x))
plt.grid()
plt.xlabel('time')
plt.ylabel('값')
plt.title('sin 함수 시각화')
plt.show()
```

```python
plt.figure( figsize = (10, 6) )
plt.plot(x, y, label='sin')
plt.plot(x, np.cos(x), label='cos')
# 범례 추가 (위에서 label 추가)
plt.legend()
plt.grid()
plt.xlabel('time')
plt.ylabel('값')
plt.title('sin 함수 시각화')
plt.show()
```

```python
plt.figure( figsize = (10, 6) )
plt.plot(x, y, label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.legend()
plt.grid()
plt.xlabel('time')
plt.ylabel('값')
plt.title('sin 함수 시각화')
# 중요한 부분 발췌(limit)
plt.xlim(0, np.pi)
plt.ylim(-1.2, 1.2)
plt.show()
```

```python
# 데코레이션
plt.figure()
# 선의 색상, 선의 모양, 데이터포인트(변곡점) 모양 지정
# dashed : 점선
# marker = 'o' : 실데이터가 있는 곳 특정 
plt.plot([1, 2, 3, 4, 5, 4, 3, 2, 1], color = 'green', linestyle='dashed', 
         marker='o', markerfacecolor = 'red', markersize = 10)
plt.show()
```
