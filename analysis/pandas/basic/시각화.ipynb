```python
# 아이리스 데이터 로드 함수
def get_iris_df() :
    ds = sklearn.datasets.load_iris()
    # 관찰결과 : data(실데이터) 항목과 feature_names(컬럼)을 획득
    # print(type(ds), ds)
    # df로 생성
    df = pd.DataFrame(ds['data'], columns = ds['feature_names'])
    
    # df 자체에 데이터는 존재하나, 그 데이터가 무엇을 의미하는지가 없다
    # 본 데이터는 지도학습법에 관련된 데이터로서 형태를 맞춰주는 것이 중요하다
    # print('데이터가 지향하는 결과 종류 : ', ds['target_names'])
    # { 0 : 'setosa', 1 : 'versicolor', ... }
    # print( dict( zip(range(3), ds['target_names']) ) )
    
    # 관련없는 2개의 연속데이터를 키와 값을 세트로 구성 : zip()
    target_dic = dict( zip(range(3), ds['target_names']) ) 
    tmp = [ target_dic[n] for n in ds['target'] ]
    df['species'] = tmp
    return df
```

```python
get_iris_df()
```

```python
df = get_iris_df()
df.head(2)
```

## 원형차트 : pie chart
- 중요한 문서, 발표자료에는 거의 사용하지 않는다.
- 그러나 데이터를 깔끔하게 보여주는 장점을 가지고 있다.
- 원형차트의 데이터는 수치표로도 보여줄 수 있다.
- 사용 이유, 시각화의 목적 : 같은 정보를 어떻게 하면 좀 더 쉽게 이해시킬 수 있을까?
- 비교하는 데이터의 종류가 적을 경우 유리

```python
# 종별 변수 데이터 합계
sums_by_species = df.groupby('species').sum()
sums_by_species
```

```python
# 첫번째 컬럼 데이터 획득
# 첫번째 컬럼
var_name = sums_by_species.columns[0]
var_name, sums_by_species[var_name]
```

```python
sums_by_species[var_name].plot(kind='pie', fontsize=10)
```

```python
sums_by_species[var_name].plot(kind='pie', fontsize=20)
# 데이터의 이름이 라벨을 가리니, 조정
plt.ylabel(var_name,  horizontalalignment='right' )
# 제목
plt.title('%s 기준으로 분류한 붓꽃의 데이터 크기 비교' % var_name)
```

```python
# 데이터 확인
sums_by_species
```

```python
# 4개 구역 -> layout(2, 2) 나눠서 차트화
# legend : 범례
sums_by_species.plot(kind = 'pie', subplots = True, layout=(2,2), legend=False)
```

## 막대그래프 : bar chart
- 원형차트와 목적은 동일
- 통상 막대그래프가 더 적합한 경우가 많다
- 여러 값을 상대적으로 크기 비교하기 적합하다

```python
plt.title('꽃받침 너비로 분류한 분꽃', fontsize=20)
sums_by_species[var_name].plot(kind='bar', fontsize=15, rot=30) # rot : rotation
```

```python
sums_by_species.plot(kind='bar', fontsize=12, subplots=True)
plt.suptitle('종별 전체 측정값 비교')
```
