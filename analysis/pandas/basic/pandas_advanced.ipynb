### 데이터프레임 병합
- concat()
- merge()

```python
import numpy as np
import pandas as pd
```

```python
# 데이터 준비
# 서로 공통된 컬럼이 1개 존재
left = pd.DataFrame({
    'key' : list('1234'),
    'A' : ['AO', 'A1', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B3']
})
right = pd.DataFrame({
    'key' : list('0123'),
    'C' : ['C0', 'C1', 'C2', 'C3'],
    'D' : ['D0', 'D1', 'D2', 'D3']
})
```

```python
left
```

```python
right
```

```python
# 합치기 : 두 개의 df에서 공통으로 존재하는 컬럼 key를 기준으로 merge() 수행
# 교집합의 맥락 : key 컬럼에 공통된 값을 가진 데이터만 존재시킴
pd.merge(left, right, on='key')
```

```python
pd.merge(right, left, on='key')
```

```python
# 교집합
# inner
pd.merge(left, right, on='key', how='inner')
```

```python
# right 기준으로 데이터를 다 맞춘다 -> left에는 존재하지 않으므로 NaN 처리
pd.merge(left, right, on='key', how='right')
```

```python
# 합집합
# outer
pd.merge(left, right, on='key', how='outer')
```

### concat

```python
# concat 대상을 리스트로 나열
# 상하로
pd.concat([left, right], sort=True)
```

```python
# 좌우로
pd.concat([left, right], sort=True, axis=1)
```
