## folium
- 파이썬을 이용해 지도 시각화를 제공하는 라이브러리
- pip install folium
- 0.10.0

```python
import folium
```

###### folium.Map( location = gps, zoom_start = zoom )

```python
# 기본
gps=[35.1685752, 129.131414]
zoom = 16
map_centom = folium.Map(location = gps, zoom_start = zoom)
map_centom
```

```python
# 지도의 타입 조정
# map_centom = folium.Map(location = gps, zoom_start = zoom, tiles = 'Stamen Toner')
map_centom = folium.Map(location = gps, zoom_start = zoom, tiles = 'Stamen Terrain')
map_centom
```

###### folium.Marker(gps, popup = '센텀신세계', icon = folium.Icon(icon = 'cloud')).add_to(map_centom)

```python
map_centom = folium.Map(location = gps, zoom_start = zoom)
folium.Marker(gps, popup = '센텀신세계', icon = folium.Icon(icon = 'cloud')).add_to(map_centom)
folium.Marker([35.1698862, 129.130424], popup = '스파월드', 
              icon = folium.Icon(icon='info-sign', color = 'red') ).add_to(map_centom)
map_centom
```

###### folium.CircleMarker(gps, radius = 50, popup = '신세계존', color = '#3186cc', fill_color = '#00ff00').add_to(map_centom)

```python
# 원형존
folium.CircleMarker(gps, radius = 50, popup = '신세계존', color = '#3186cc', fill_color = '#00ff00').add_to(map_centom)
map_centom
```

###### folium.RegularPolygonMarker( [35.1698862, 129.130424], fill_color = '#00ff00', number_of_sides = 3, radius = 50).add_to(map_centom)

```python
# 다각형 -> 3, 4, 6, 8, ...
folium.RegularPolygonMarker( [35.1698862, 129.130424], fill_color = '#00ff00', number_of_sides = 3, radius = 50).add_to(map_centom)
map_centom
```
